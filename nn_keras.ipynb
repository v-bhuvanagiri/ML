```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequentialtial
```

```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Vaishnavi\Downloads\Bank_Personal_Loan_Modelling.csv")
```

```python
x= data.drop(["ID","ZIP Code","CreditCard"],axis = 1).values
y = data["CreditCard"].values
```

```python
x.shape
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()
x= norm.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state= 6)
```

```python
model1 =Sequential([
    Dense(64,input_dim = 11,activation = "relu"),
    Dense(32,activation="relu"),
    Dense(1,activation="relu")
])
```

```python
model1.compile(loss = "binary_crossentropy",optimizer = "adam",metrics = ["accuracy"])
```

```python
history = model1.fit(x_train,y_train,batch_size = 32, epochs = 100,validation_data = (x_test,y_test))
```

```python
test_loss,test_acc = model1.evaluate(x_test,y_test)
print(f"Test loss : {test_loss:.4f},Tes Accuracy : {test_acc:.4f}")
```

```python
print(history.history.keys())
```
