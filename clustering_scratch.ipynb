```python
import numpy as np
import random
```

```python
def intialize_centroids(x,k):
    n_samples,n_features = x.shape
    #print(n_samples)
    #print(n_features)
    centroids = np.zeros((k,n_features))
    #print(centroids)
    for i in range(k):
        centroid = x[np.random.choice(range(n_samples))]
        centroids[i] = centroid
    return centroids    
```

```python
intialize_centroids(x,5)
```

```python
def closest_centroids(x,centroids):
    distance = np.sqrt((x - centroids[:,np.newaxis])**2).sum(axis=2)
    return np.argmin(distance,axis=0)
```

```python
def update_centroids(x,labels,k):
    centroids = np.zeros((k,x.shape[1]))
    for i in range(k):
        centroids[i] = x[labels ==i].mean(axis=0)
    return centroids
```

```python
def kmeans(x,k,max_iterations):
    centroids = intialize_centroids(x,k)
    for i in range(max_iterations):
        labels = closest_centroids(x,centroids)
        centroids = update_centroids(x,labels,k)
    return centroids,labels    
```

```python
centroids,labels = kmeans(x,5,100)
```
